# Maplotlib basic

## pyplot : 그림 판때기라 생각 (객체)
- figure라 하는 그림(여러 그래프들)을 쌓아 올려서 표현하는 것이 maplotlib
- 고정된 argument가 없어서 parameter로 받아서 직접 찾아야함(구글링 필요)

```python
import matplotlib.pyplot as plt

X = range(100)
Y = range(100)
plt.plot(X, Y)
plt.show()
```

```python
import numpy as np 

X_1 = range(100)
Y_1 = [np.cos(value) for value in X]
X_2 = range(100)
Y_2 = [np.sin(value) for value in X]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.plot(range(1), range(1))
plt.show()
```

### Figure & Subplot


### subplot은 그리드개념으로 접근하면 이해하기 쉬움
> 칸을 쪼개고 그 칸에서 몇 번쨰에 그림을 그릴 것인가

```python
fig = plt.figure() # figure 반환
fig.set_size_inches(10,3) # 크기지정 (가로, 세로)

ax_1 = fig.add_subplot(1,2,1) # 두개의 plot 생성 - 1행 2열의 subplot 중 1번째

ax_2 = fig.add_subplot(1,2,2)  # 두개의 plot 생성 - 1행 2열의 subplot 중 2번째

ax_1.plot(X_1, Y_1, c="b")  # 첫번째 plot을 그림
ax_2.plot(X_2, Y_2, c="g")  # 두번째 plot을 그림
plt.show() # show & flush
```

```python
fig = plt.figure() # figure 반환
fig.set_size_inches(10,10) # 크기지정

ax_1 = fig.add_subplot(2,2,1) # 두개의 plot 생성
ax_2 = fig.add_subplot(2,2,2)  # 두개의 plot 생성
ax_3 = fig.add_subplot(2,1,2)  # 두개의 plot 생성

ax_1.plot(X_1, Y_1, c="b")  # 첫번째 plot
# ax_2.plot(X_2, Y_2, c="g")  # 두번째 plot
ax_3.plot(X_1, Y_1, c="k")
plt.show() # show & flush
```

```python
fig = plt.figure()
fig.set_size_inches(10,10) # 싸이즈 설정
plt.style.use("ggplot")    # 스타일적용
plt.title('title')

ax = []
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
for i in range(1,7):
    ax.append(fig.add_subplot(2,3,i)) # 두개의 plot 생성
    X_1 = np.arange(50)
    Y_1 = np.random.rand(50)
    c=colors[np.random.randint(1, len(colors))]
    
    ax[i-1].plot(X_1, Y_1, c=c)
```

### 각종 기능

#### 제목
`plt.title('')`
#### 범례
`plt.legend('')`
#### grid 설정
`plt.grid('')`
#### 축 limit 설정
`plt.xlim(0, 100)`  
`plt.ylim(-100, 100)`
#### 저장 (plt.show를 하기 전에 해야함 - 나오는 순간 객체가 사라짐)
`plt.savefig('filename')`

```python
fig = plt.figure() # figure 반환
fig.set_size_inches(10,10) # 크기지정

ax_1 = fig.add_subplot(2,2,1) # 두개의 plot 생성
ax_2 = fig.add_subplot(2,2,2)  # 두개의 plot 생성
ax_3 = fig.add_subplot(2,1,2)  # 두개의 plot 생성

ax_1.plot(X_1, Y_1, c="b")  # 첫번째 plot
ax_2.plot(X_2, Y_2, c="g")  # 두번째 plot
ax_3.plot(X_1, Y_1)

ax_2.margins(2, 2)           # Values >0.0 zoom out

left  = 0.125  # the left side of the subplots of the figure
right = 0.9    # the right side of the subplots of the figure
bottom = 0.1   # the bottom of the subplots of the figure
top = 0.9      # the top of the subplots of the figure
wspace = 0.5   # the amount of width reserved for blank space between subplots
hspace = 0.1   # the amount of height reserved for white space between subplots
fig.subplots_adjust(left, bottom, right, top, wspace, hspace=None)
# fig.tight_layout()


plt.show() # show & flush
```

```python
fig = plt.figure()
fig.set_size_inches(10,10) # 싸이즈 설정
ax= fig.subplots(ncols=3, nrows=3) 

colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
for i in range(3):
    for j in range(3):
        X_1 = np.arange(50)
        Y_1 = np.random.rand(50)
        c=colors[np.random.randint(1, len(colors))]

        ax[i][j].plot(X_1, Y_1, c=c)
plt.show()
```

### Examples
#### 월별 통화량 데이터 총합 구해보기

```python
import pandas as pd
df = pd.read_csv("../data/phone_data.csv")
df.head()
```

```python
result = df.groupby(["month"])["duration"].sum()
result
```

```python
plt.plot(result.index, result, color="b", linestyle="dashed")
plt.title("Durations per month")
plt.show()
```

#### 일별 네트워크별 데이터의 총량을 다중 plot으로 작성하라`

```python
def parsedate(date_info):
    # import datetime
    return date_info[:9]
```

```python
df["day"] = df["date"].map(parsedate)
df.head()
```

```python
result = df.groupby(["network", "day"])["duration"].sum().reset_index()
result.head()
```

```python
result["network"].unique().tolist()
```

```python
result[result["network"] == "data"].head()
```

```python
fig = plt.figure()
fig.set_size_inches(10,10) # 싸이즈 설정
# plt.style.use("ggplot")    # 스타일적용

network_types = result["network"].unique().tolist()
ax = []
for i in range(1,7):
    ax.append(fig.add_subplot(2,3,i)) # 두개의 plot 생성
    network_name = network_types[i-1]
    
    plt.title(network_name)

    X_1 = result[result["network"] == network_name]["day"]
    Y_1 = result[result["network"] == network_name]["duration"]
    ax[i-1].get_xaxis().set_visible(False)
    ax[i-1].plot(X_1, Y_1)
```
