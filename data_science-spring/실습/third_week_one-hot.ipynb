# Polynimial Term 추가

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
classification_df = pd.read_csv('https://raw.githubusercontent.com/ralbu85/DataScience_2022S/master/data/classification_data.csv')
classification_df.head()
```

```python
# 독립변수 추리기
inde_variable = classification_df[['X1', 'X2']]
# 몇차식으로 할 건지, include_bias : 편향을 넣을건지, interaction_only : 2차식 빠진다
poly = PolynomialFeatures(2, include_bias= False, interaction_only=True)
poly.fit_transform(inde_variable)
```

```python

poly.get_feature_names(inde_variable.columns)
```

```python
pd.DataFrame(poly.fit_transform(inde_variable), columns = poly.get_feature_names(inde_variable.columns))
```

# 타이타닉 데이터 전처리

```python
from sklearn.datasets import fetch_openml
```

```python
X, y = fetch_openml("titanic", version=1, as_frame=True, return_X_y=True)
```

```python
# 승선자에대한 정보
X
```

```python
# 생존 정보
y
```

#### 분석에 필요할만한 column 추리기
- `nunique()`
  - 너무 많은 종류에 데이터가 있는 column은 날린다
- `isnull().sum()`
  - 각 column별 null 값을 확인하여 너무 많으면 날린다

```python
# 각 열별로 고유한 값의 개수가 몇 개인지
X.nunique()
```

```python
columns = ['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked']
X = X[columns]
X
```

```python
X = X.dropna(axis= 0)
X
```

```python
X = pd.get_dummies(X, columns=['pclass', 'sex', 'embarked'])
X
```

```python
# 성별은 남자이거나 여자이기 때문에 한 개만 있어도 됨
X = X.drop(columns='sex_male')
X
```

#### drop된 Index를 y에도 맞추기

```python
X.index
```

```python
# 새로운 df에 저장
df = X
# series 형태의 y를 x.index를 통해 뽑아 내고 새로운 데이터프레임에 추가
df['survived'] = y[X.index]
```
